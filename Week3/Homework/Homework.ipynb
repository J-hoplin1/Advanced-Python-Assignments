### Q1: String Extraction Puzzle - Is there an Error?
```python
def second(thelist):
    start=thelist.index(',')
    tail = thelist[start+1:]
    end = tail.index(',')
    result =tail[:end]
    return result
```
### A1 : "E: There is no error"

- 우선 `list`의 .index()메소드는 주어진 문자가 최초로 존재하는 index를 반환하게 됩니다. 그렇기 때문에, start변수에는 처음 만나는 index인 `index 1`이 들어갈 것입니다. tail은 ','의 다음문자인 B부터 슬라이싱을 하는 부분입니다. 아래 실행 결과에서 보듯이, tail변수는 'B,C'가 나오게 되고, 이 상태에서 다시 ',' 를 기준으로 왼쪽으로 슬라이싱을 하면, 가운데 문자열인 'B'가 나오게 됩니다. 슬라이싱을 위해 기준 변수인 `end`에 index값을 넣어주게 되는것입니다. 리스트를 슬라이싱 할때 \[:x\] 형태로 하게되면, 해당 리스트의 처음부터 x-1 index까지 슬라이싱 하게 됩니다. 즉, 결론적으로 result에는 가운데 문자열인 'B'만 남게되는것입니다.

````python
def second(thelist):
    start=thelist.index(',')
    print(f"start : {start}")
    tail = thelist[start+1:]
    print(f"tail : {tail}")
    end = tail.index(',')
    print(f"end : {end}")
    result =tail[:end]
    print(f"result : {result}")
    return result

second('A,B,C')
````

### Q9 : Swap List Values? - What gets printed?

```python
def swap(b,h,k):
    tmp = x[a]
    x[a] = x[b]
    x[b] = tmp

x = [5,4,7,6,8]
swap(x,3,4)
print(x[3])
```
### A9 : "A: 8"

- swap()함수의 기능을 우선 보면, Sequence타입의 b와 그 후 index를 의미하는 h,k를 입력으로 받은 후 b의 index h,k값을 서로 바꿔주는 함수입니다. 여기서 알아야 할것은, 리스트는 객체타입이므로, Call by reference가 되므로, 함수에 값으로 넘겼을때 원본값이 영향을 받는다는것입니다. 그렇기 때문에, swap()함수를 거친 이후에는 index 3, 4번의 값이 서로 바뀐 값인 [5,4,7,8,6]이 될것이며 그에 따라 index 3의 값은 8이 됩니다



````python
def swap(x,a,b):
    tmp = x[a]
    x[a] = x[b]
    x[b] = tmp

x = [5,4,7,6,8]
swap(x,3,4)
x[3]
````
